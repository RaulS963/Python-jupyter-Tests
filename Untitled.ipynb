```python
class Person:
    childrenList =[]
    def __init__(self,name=None,age=None,sex=None,country=None):
        self.name = name
        self.age = age
        self.sex = sex
        self.country = country
        
    def parentof(self,child=None):
        if child not in self.childrenList:
            self.childrenList.append(child.getname())
            
        
    def getChildrenList(self):
        return self.childrenList
    
    def getname(self):
        return self.name
    
    def info(self):
        s = f"name: {self.name}; age: {self.age} ; sex: {self.sex} ; country: {self.country}"
        return s
        
           
```

```python
raj = Person('raj',30)
raj.info()

sumit = Person('sumit',6,'m')
priya = Person('priya',2,'f')

raj.parentof(sumit)
raj.parentof(priya)

raj.getChildrenList()
```
